## **Nettoyage et harmonisation des données**
### **Harmonisation des variables de nos deux datasets**
Nous allons regarder les colonnes uniques de chacun de ces deux datasets, puis harmoniser les noms et types de colonnes afin de pouvoir les regrouper. Les colonnes trop peu remplies ou non nécessaires à ce projet seront enlevées.

```python
data2015.columns
```

```python
data2016.columns
```

```python
column_2015 = data2015.columns
column_2016 = data2016.columns
common_columns = column_2015.intersection(column_2016)
unique_2015 = column_2015.difference(common_columns)
unique_2015
```

```python
unique_2016 = column_2016.difference(common_columns)
unique_2016
```

```python
unique2015df = data2015[unique_2015]
unique2015df
```

```python
unique2016df = data2016[unique_2016]
unique2016df
```

```python
unique2015df.info()
```

```python
unique2016df.info()
```

10 colonnes unique pour 2015 et 9 pour 2016. Nous voyons notamment des colonnes de commentaires, vides ou presque, observons ce qu'il y a dedans.

```python
unique2016df['Comments'].unique()
```

```python
unique2015df['Comment'].unique()
```

Nous allons supprimmer les colonnes de commentaires pour chaque dataset car ce n'est pas utile pour notre analyse. 

```python
data_2015_updated=data2015.copy()
data_2015_updated.drop('Comment', axis=1,inplace=True)

data_2016_updated=data2016.copy()
data_2016_updated.drop('Comments', axis=1,inplace=True)
```

Nous harmonisons les types et nom de colonne ZipCodes afin de pouvoir les combiner par la suite.

```python
data_2015_updated.rename({'Zip Codes': 'ZipCode'}, axis=1, inplace=True)
data_2015_updated['ZipCode']=data_2015_updated['ZipCode'].astype('object')
data_2016_updated['ZipCode']=data_2016_updated['ZipCode'].astype('object')
```

Nous extrayons de la colonne 'Location' les Latitudes et Longitudes pour les données de 2015.

```python
import ast

l=[]
for row in data_2015_updated['Location']:
    a=ast.literal_eval(row)
    l.append(a['latitude'])

data_2015_updated['Latitude']=l

l1=[]
for row in data_2015_updated['Location']:
    a=ast.literal_eval(row)
    l1.append(a['longitude'])

data_2015_updated['Longitude']=l1

data_2015_updated.drop('Location', axis=1, inplace=True)
```

Certaines variables de ce dataset de la ville de Seattle ne nous apportent pas d'informations concernant la consommation énergétique. Nous suprimons donc ces colonnes pour notre étude.

```python
data_2015_updated.drop(['OtherFuelUse(kBtu)','2010 Census Tracts','SPD Beats','City Council Districts','Seattle Police Department Micro Community Policing Plan Areas'], axis=1, inplace=True)
```

```python
data_2016_updated.drop(['City','Address','State'], axis=1, inplace=True)
```

Nous renommons les colonnes concernant les GHGEmissions des données de 2015 pour qu'elles aient le même nom que les données de 2016.

```python
data_2015_updated.rename({'GHGEmissions(MetricTonsCO2e)':'TotalGHGEmissions', 'GHGEmissionsIntensity(kgCO2e/ft2)': 'GHGEmissionsIntensity'}, axis=1, inplace=True)
```

Regardons maintenant si il reste des colonnes uniques :

```python
column_2015_updated = data_2015_updated.columns
column_2016_updated = data_2016_updated.columns
common_columns = column_2015_updated.intersection(column_2016_updated)
unique_2015_updated= column_2015_updated.difference(common_columns)
unique_2015_updated
```

```python
unique_2016_updated= column_2016_updated.difference(common_columns)
unique_2016_updated
```

Plus de colonnes uniques, nous pouvons maintenant combiner les deux datasets. Vérifions que les types de données correspondent. 

```python
data_2015_updated.info()
```

```python
data_2016_updated.info()
```

Modifions les types de données pour qu'ils soient identiques pour chaque dataset afin de pouvoir les combiner.

```python
data_2015_updated[['Latitude', 'Longitude', 'NumberofBuildings']].astype(float)
data_2016_updated['NumberofFloors'].astype(float)
data_2015_updated.drop('DefaultData', axis=1, inplace=True)
data_2016_updated.drop('DefaultData', axis=1, inplace=True)
```

Nous pouvons désormais combiner les deux jeux de données. Puis ensuite nous ne gardons qu'une seule ligne par bâtiment grâce à l'**OSEBuildingID**. Lorsque pour un bâtiment nous avons 2 lignes de données nous garderons les données les plus récentes, à savoir celles de 2016.

```python
seattle_data = pd.concat([data_2015_updated, data_2016_updated], join='outer', ignore_index=True)

seattle_data_uniques = seattle_data.drop_duplicates(subset=['OSEBuildingID'],keep='last').sort_values(by=['OSEBuildingID']).reset_index(drop=True)
seattle_data_uniques
```

```python
seattle_data_uniques.columns
```

```python
seattle_data_uniques = seattle_data_uniques.drop_duplicates(['TaxParcelIdentificationNumber','Latitude', 'Longitude'])
```

Nous travaillons maintenant sur un grand dataset que nous avons nommé ici **seattle_data_uniques** dans lequel nous avons supprimé les doublons qui avaient les mêmes **'TaxParcelIdentificationNumber', 'Latitude' & 'Longitude'**.

### **Analyse du remplissage des variables**

Regardons maintenant parmi toutes les variables à notre disposition, comment celles-cis sont renseignées. Nous allons regarder leur remplissage. 

```python
missing_values = seattle_data_uniques.isna().sum()

plt.figure(figsize=(20,5))
plt.title('Valeurs manquantes par variable')
missing_values.sort_values().plot(kind='bar')
plt.xticks(rotation=30,ha="right")
plt.tight_layout()
plt.show()
```

Nous voyons ici le nombre de valeurs manquantes par colonne. Il y a quelques colonnes avec beaucoup de valeurs manquantes (>1000). Nous supprimons les colonnes ayant moins de 2400 valeurs remplies.

```python
seattle_data_uniques.dropna(axis=1, thresh=2400, inplace=True)
```

Maintenant que nous avons des colonnes remplies, essayons d'évaluer leur pertinence pour notre étude.

## Analyses Univariées et multivariées

Afin de mieux comprendre nos variables et de remplir les données manquantes de notre dataset, nous allons réaliser quelques analyses simples sur nos données.
### Site Energy Use, Electricity, GHG Emissions
Ces variables sont ici au coeur du projet puisque qu'elles nous transmettent les informations énergétiques des bâtiments de la ville de Seattle. Regardons comment celles-ci sont distribuées, puis quelle est leur corrélation. Cela nous permettra alors de nous focaliser sur une seule d'entres elles pour la suite de l'étude, notamment pour la partie prédiction. 

```python
sns.displot(seattle_data_uniques, x='SiteEnergyUse(kBtu)', aspect=4)
plt.xlim(0,50000000)
```

Nous constatons ici que la casi totalité des valeurs se trouvent en dessous de 3x10^7 kBtu.

```python
sns.displot(seattle_data_uniques, x='Electricity(kWh)', aspect=4)
plt.xlim(0,10000000)
```

Nous remarquons que la grande majorité des bâtiments ont une consommation électrique inférieure à 4x10^6 kW/h.

```python
sns.displot(seattle_data_uniques, x='GHGEmissionsIntensity', aspect=4)
plt.xlim(0,10)
```

Nous observons ici que la plupart des bâtiments ont une intensité d'émissions de gaz à effet de serre inférieure à 4 kgCO2e/ft2.

```python
sns.displot(seattle_data_uniques, x='TotalGHGEmissions', aspect=4)
plt.xlim(0,1000)
```

La majorité des bâtiments ont des émissions totales de gaz à effet de serre inférieure à 500 MetricTonsCO2e.

```python
sns.regplot(y='SiteEnergyUse(kBtu)' ,x='ENERGYSTARScore',  data=seattle_data_uniques, line_kws={'color':'red'})
plt.ylim(0,30000000)
```

```python
sns.regplot(y='Electricity(kWh)' ,x='ENERGYSTARScore',  data=seattle_data_uniques, line_kws={'color':'red'})
plt.ylim(0,4000000)
```

```python
sns.regplot(x='ENERGYSTARScore' ,y='TotalGHGEmissions',  data=seattle_data_uniques, line_kws={'color':'red'})
plt.ylim(0,500)
```

Lorsque la consommation d'électricité ou les GHG Emissions augmentent, l'Energy Star Score diminue.

```python
sns.regplot(y='SiteEnergyUse(kBtu)' ,x='TotalGHGEmissions',  data=seattle_data_uniques, line_kws={'color':'red'})
plt.xlim(0, 500)
plt.ylim(0, 30000000)
```

```python
sns.regplot(y='Electricity(kWh)' ,x='TotalGHGEmissions',  data=seattle_data_uniques, line_kws={'color':'red'})
plt.xlim(0, 500)
plt.ylim(0, 4000000)
```

La consommation électrique et l'utilisation énergique du site augmentent en même temps que les GHG Emissions.

Regardons maintenant comment toutes ces variables sont corrélées grâce à une heatmap. 

```python
seattle_corr = seattle_data_uniques[['SiteEnergyUse(kBtu)', 'SiteEnergyUseWN(kBtu)','Electricity(kWh)', 'Electricity(kBtu)','TotalGHGEmissions']]

corrMatrix = seattle_corr.corr()

plt.figure(figsize=(7,7))
plt.title("Matrice de corrélation", fontsize=20)
sns.heatmap(corrMatrix,fmt= '.2f', annot=True, cbar=True)
plt.xticks(rotation=30)
plt.show()
```

Nous observons que la variable SiteEnergyUse(kBtu) est celle qui est la plus corrélée à toutes les autres (au moins 86%). Nous allons donc l'utiliser pour la suite de nos analyses multivariées.
